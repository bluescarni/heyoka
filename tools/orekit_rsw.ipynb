# Initialisation

```python
import orekit
vm = orekit.initVM()

from org.orekit.time import TimeScalesFactory, AbsoluteDate
from org.orekit.frames import FramesFactory, LocalOrbitalFrame, LOFType
from org.hipparchus.geometry.euclidean.threed import Vector3D
from org.orekit.utils import PVCoordinates
from org.orekit.orbits import CartesianOrbit
```

# Keplerian orbit

```python
# Initial setup of several constants.
tai = TimeScalesFactory.getTAI()
# NOTE: don't think this initial date matters, but it is still required.
initial_date = AbsoluteDate(2008, 10, 1, 0, 0, 00.000, tai)
mu =  3.986004415e+14

# Setup of the inertial frame.
inertial_frame = FramesFactory.getEME2000()

# State of the chief in the interial frame.
chief_pos = Vector3D(1.9990152502378467e+06,  -4.2466313738494366e+05, 6.7714722017919971e+06)
chief_vel = Vector3D(-6.9397802817958964e+03, -2.1318724003511638e+03, 1.9205549571233923e+03)
chief_pv = PVCoordinates(chief_pos, chief_vel)

# Orbit of the chief in the inertial frame. This specifies
# that the chief is moving in a Keplerian orbit.
chief_orbit = CartesianOrbit(chief_pv, inertial_frame, initial_date, mu)

# Chief's RSW frame. Orekit calls this the QSW frame.
local_frame = LocalOrbitalFrame(inertial_frame, LOFType.QSW, chief_orbit, "qswFrame")

# Define the transformation from inertial to QSW frame.
inertial_to_local = inertial_frame.getTransformTo(local_frame, initial_date)

# Inertial state vector of the deputy.
deputy_pos = Vector3D(1.9901598679710650e+06,  -4.3171886347018217e+05, 6.7720686840924025e+06)
deputy_vel = Vector3D(-6.9420492298495501e+03, -2.1395918103707527e+03, 1.9120431913109790e+03)

# Transform the deputy state from inertial to RSW.
inertial_to_local.transformPVCoordinates(PVCoordinates(deputy_pos, deputy_vel))
```

# Inertial version

```python
local_inertial_frame = LocalOrbitalFrame(inertial_frame, LOFType.QSW_INERTIAL, chief_orbit, "qswInertialFrame")

inertial_to_local_inertial = inertial_frame.getTransformTo(local_inertial_frame, initial_date)

inertial_to_local_inertial.transformPVCoordinates(PVCoordinates(deputy_pos, deputy_vel))
```
